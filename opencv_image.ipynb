# 00: Opening an Image

```python
import cv2
from matplotlib import pyplot as plt
image_file = "data/test1.png"
img = cv2.imread(image_file)
```

```python
cv2.imshow("original Image", img) 
cv2.waitKey(0) # This will crash the program to open the image so below is the appropriate way
```

```python
# Displaying different images with actual size in matplotlib subplot
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_image_in_actual_size(im_path):
    dpi = 80
    im_data = mpimg.imread(im_path)

    if im_data.ndim == 2:
        height, width = im_data.shape
    else:
        height, width, _ = im_data.shape

    figsize = width / float(dpi), height / float(dpi)
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])
    ax.axis('off')
    ax.imshow(im_data, cmap='gray' if im_data.ndim == 2 else None)
    plt.show()
```

```python
display_image_in_actual_size(image_file)
```

```python
inverted_image = cv2.bitwise_not(img)
cv2.imwrite("temp/inverted.png", inverted_image)
```

```python
display_image_in_actual_size("temp/inverted.png")
```

# 03: Binarization

### -> Convert an Image into black & white

```python
def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
```

```python
gray_image = grayscale(img)
cv2.imwrite("temp/gray.png", gray_image)
```

```python
display_image_in_actual_size("temp/gray.png")
```

```python
# Changing the color space to gray and then applying thresholding to convert to binary image (Black and White)
thresh, im_bw = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
cv2.imwrite("temp/bw.png", im_bw)
```

```python
display_image_in_actual_size("temp/bw.png")
```
